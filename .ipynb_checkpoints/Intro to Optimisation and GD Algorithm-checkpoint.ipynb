## Example 2 - Multiple Minima vs Initial Guess & Advanced Functions

$$g(x) = x^4 - 4x^2 + 5$$

```python
x_2 = np.linspace(-2, 2, 1000)
```

```python
def g(x):
    return (x**4) - (4* x**2) + 5

def dg(x):
    return (4 * x**3) - 8 * x
```

```python
plt.figure(figsize = (10, 5))

plt.subplot(1,2,1)

plt.title("Cost function")
plt.xlabel('X', fontsize = 15)
plt.ylabel('g(x)', fontsize = 15)
plt.xlim(-2, 2)
plt.ylim(0,6)

plt.style.use('fivethirtyeight')

plt.plot(x_2, g(x_2), lw = 2, alpha = .8)

plt.subplot(1,2,2)

plt.title("Slope of the cost function")
plt.xlabel('X', fontsize = 15)
plt.ylabel('dg(x)', fontsize = 15)
plt.xlim(-2, 2)
#plt.ylim(0,8)

plt.style.use('fivethirtyeight')

plt.plot(x_2, dg(x_2), color = 'red', lw = 2, alpha = .8)
```


## Gradient Descent as a Python Function

```python
def gradient_descent(derivative_func, initial_guess, step_multiplier = 0.02, precision = 0.001, max_iter = 300):
    
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for i in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x -= step_multiplier * gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, 1.9, 0.02, 0.001)

print('Local min occurs at:', local_min)
print('Number of steps:', len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func = dg, initial_guess = -0.1)

print('Local min occurs at:', local_min)
print('Number of steps:', len(list_x))
```

```python
plt.figure(figsize = (18, 5))

plt.subplot(1,3,1)

plt.title("Cost function")
plt.xlabel('X', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)
plt.xlim(-2,2)
plt.ylim(-1,6)

plt.style.use('fivethirtyeight')

plt.plot(x_2, g(x_2), lw = 2, alpha = .8)
plt.scatter(list_x, g(np.array(list_x)), color = 'r', s= 100, alpha = 0.6)

plt.subplot(1,3,2)

plt.title("Slope of the cost function")
plt.xlabel('X', fontsize = 15)
plt.ylabel('dg(x)', fontsize = 15)
plt.xlim(-2, 2)
#plt.ylim(0,8)

plt.style.use('fivethirtyeight')

plt.plot(x_2, dg(x_2), color = 'skyblue', lw = 2, alpha = .8)
plt.scatter(list_x, deriv_list, color = 'r', s = 100, alpha = 0.5)

plt.subplot(1,3,3)

plt.title("Closeup of Slope")
plt.xlabel('X', fontsize = 15)
plt.ylabel('dg(x)', fontsize = 15)
plt.xlim(-1.5, 0)
plt.ylim(-0.5,5.5)

plt.style.use('fivethirtyeight')

plt.plot(x_2, dg(x_2), color = 'skyblue', lw = 2, alpha = .8)
plt.scatter(list_x, deriv_list, color = 'red', alpha = .7, s = 100)
```

# Example 4 - Data Viz with 3D Charts

## Minimise $$f(x,y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x,y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$ 

```python
def f(x,y):
    r = 3 ** (-x**2 - y**2)
    return 1/ (r + 1)
```

```python
x_4 = np.linspace(-2, 2, 200)
y_4 = np.linspace(-2, 2, 200)

x_4, y_4 = np.meshgrid(x_4, y_4)

x_4.shape
```

```python
# Generating 3D Plot

fig = plt.figure(figsize = (16, 12))
ax = fig.gca(projection = '3d')

ax.set_xlabel('X', fontsize = 20)
ax.set_ylabel('Y', fontsize = 20)
ax.set_zlabel('f(x, y): Cost function', fontsize = 20)

ax.plot_surface(x_4, y_4, f(x_4, y_4), cmap = cm.turbo_r, alpha = .4)
```

## Partial Derivatives and Symbolic Computation

### $$\frac{\partial f}{\partial x} = \frac{2x\ln{(3)} \cdot 3^{-x^2 - y^2}}{\left(3^{-x^2 - y^2} + 1\right)^2} $$

### $$\frac{\partial f}{\partial y} = \frac{2y\ln{(3)} \cdot 3^{-x^2 - y^2}}{\left(3^{-x^2 - y^2} + 1\right)^2} $$


```python
a, b = symbols('x, y')
print('Cost function: ', f(a, b))
```

```python
diff(f(a, b), a) # Deriv of f(x, y) wrt x
```

```python
diff(f(a,b), b) # Deriv of f(x, y) wrt y
```

```python
f(a, b).evalf(subs = {a: 1.8, b:1.0}) # Value of f(x,y) @ x = 1.8, y = 1.0
```

```python
diff(f(a, b), a).evalf(subs = {a:1.8, b:1.0}) # Partial deriv of the function wrt x @ x = 1.8, y = 1.0
```

## Batch Gradient Descent with SymPy

```python
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) #initial guess

for n in range(max_iter):
    gradient_x = diff(f(a, b), a).evalf(subs = {a: params[0], b: params[1]})
    gradient_y = diff(f(a, b), b).evalf(subs = {a: params[0], b: params[1]})
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients
    
print('Gradient Array: ', gradients)
print('X-value at minimum: ', params[0])
print('Y-value at minimum: ', params[1])
print('Cost: ', f(a,b).evalf(subs = {a:params[0], b: params[1]}))
```

```python
def fpx(x, y):
    r = 3 ** (-x**2 - y**2)
    num = 2 * x * r * np.log(3)
    den = (r + 1)**2
    return num/den

def fpy(x, y):
    r = 3 ** (-x**2 - y**2)
    num = 2 * y * r * np.log(3)
    den = (r + 1)**2
    return num/den
```

```python
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) #initial guess

for n in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients
    
print('Gradient Array: ', gradients)
print('X-value at minimum: ', params[0])
print('Y-value at minimum: ', params[1])
print('Cost: ', f(a,b).evalf(subs = {a:params[0], b: params[1]}))
```

# Graphing 3D Gradient Descent & Advanced Numpy Arrays

```python
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) #initial guess
values_array = params.reshape(1,2)


for n in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients
    #values_array = np.append(values_array, params.reshape(1,2), axis = 0)
    values_array = np.concatenate((values_array, params.reshape(1,2)), axis = 0 )
print('X-value at minimum: ', params[0])
print('Y-value at minimum: ', params[1])
```

# Example 5 - Working with Data & a Real Cost Function

## Mean Squared Error: a cost function for regression problems

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - \hat{y} \big)^2 $$

```python
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(-1, 1)
x_5.shape
```

```python
regr = LinearRegression()
regr.fit(x_5, y_5)
```

```python
regr.intercept_
```

```python
regr.coef_
```

```python
plt.scatter(x_5, y_5, s = 50)
plt.plot(x_5, regr.predict(x_5), lw = 3, color = 'orange')
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
y_hat = regr.intercept_ + regr.coef_ * x_5

y_hat
```

```python
def mse(y, yhat):
    return np.average((y - yhat)**2)
```

```python
mse(y_5, y_hat)
```

```python
mean_squared_error(y_5, y_hat)
```

## MSE & Gradient Descent

```python
def grad(x, y, thetas):
    n = y.size
    
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)
    
    return np.array([theta0_slope[0], theta1_slope[0]])
```

```python
multiplier = 0.01
n_iter = 1000
thetas = np.array([2.9, 2.9])

plot_vals = thetas.reshape(1,2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(n_iter):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)
    
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1,2)), axis = 0)
    mse_vals = np.append(mse_vals, mse(y_5, thetas[0] + thetas[1]*x_5))
    
```

```python
fig = plt.figure(figsize = (16, 12))

ax = plt.axes(projection = '3d')

ax.set_xlabel('Theta 0', fontsize = 20)
ax.set_ylabel('Theta 1', fontsize = 20)
ax.set_zlabel('Cost - MSE', fontsize = 20)

ax.plot_surface(plot_t0, plot_t1, plot_cost, cmap = 'rainbow', alpha = .4)
ax.scatter(plot_vals[:,0], plot_vals[:,1], mse_vals, s = 80, color = 'black')

plt.show()
```
